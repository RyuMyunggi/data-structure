## Array: 배열
* 같은 타입의 변수들로 이루어진 유한 집합(크기가 정해짐)
* 배열을 구성하는 각각의 값을 배열요소라고 하며 배열에서 위치를 가리키는 숫자는 인덱스(index)라고 함(리스트와의 가장 큰 차이점. 리스트에는 인덱스가 없음)
* 파이썬에서는 리스트 타입이 배열 기능을 제공
* 같은 종류의 데이터를 효율적으로 관리하는 경우에 사용
* 같은 종류의 데이터를 순차적으로 저장
* 배열이 생성 될 때 항상 크기가 주어짐
* 할당 된 메모리 공간은 연속적. 즉 배열의 아이템들은 메모리 상에서 이어져서 배치됨. 각 아이템에 인덱스를 할당할 수도 있고 인덱스를 활용하여 쉽고 직접적으로 아이템에 접근 가능
* 배열의 모든 아이템들은 똑같은 사이즈를 가짐
* 파이썬에서는 리스트 타입이 배열 기능을 제공

## 배열의 장단점
### 장점
* 빠른 접근이 가능
### 단점

```python
## 1차원 배열

array = [1, 2, 3, 4, 5]
print(array)
```

```python
## 2차원 배열
array = [['a', 'b', 'c'], ['d', 'e', 'f']]
print(array)
print(array[0])
print(array[1])
```
